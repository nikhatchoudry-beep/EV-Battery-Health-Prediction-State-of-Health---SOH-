```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
```

```python
import pandas as pd
data=pd.read_csv("battery_dataset.csv")
data.head()
```

```python
data=data[['Voltage','Current','Temperature','Capacity','Cycle','SOH']]
data.head()
```

```python
x=data[['Voltage','Current','Temperature','Capacity','Cycle']]
y=data['SOH']
```

```python
x.head()
```

```python
y.head()
```

```python
x_train, x_test, y_train, y_test= train_test_split(x,y, test_size=0.2,random_state=42)
```

```python
model = RandomForestRegressor()
model.fit(x_train, y_train)
```

```python
score= model.score(x_test, y_test)
print("Model Accuracy:",score)
```

```python
new_data = [[3.7,1.5,28,2.5,120]]
prediction = model.predict(new_data)
print("Predicted Battery SOH:",prediction[0])
```

```python
import pandas as pd

new_data = pd.DataFrame([[3.6,0.9,25,2.5,1]],
columns=['Voltage','Current','Temperature','Capacity','Cycle'])

prediction = model.predict(new_data)

print("Predicted Battery SOH:", prediction[0])
```
